# eBay Kleinanzeigen analysis

Project to clean the data and analyze the used car listings.

```python
import pandas as pd
import numpy
```

```python
autos = pd.read_csv('autos.csv', encoding='Latin-1')
```

```python
autos.rename({'yearOfRegistration': 'registration_year'}, inplace=True, axis=1)
autos.rename({'monthOfRegistration': 'registration_month'}, inplace=True, axis=1)
autos.rename({'notRepairedDamage': 'unrepaired_damage'}, inplace=True, axis=1)
autos.rename({'dateCreated': 'ad_created'}, inplace=True, axis=1)
autos.rename({'yearOfRegistration': 'registration_year'}, inplace=True, axis=1)
autos.rename({'dateCrawled': 'date_crawled'}, inplace=True, axis=1)
autos.rename({'offerType': 'offer_type'}, inplace=True, axis=1)
autos.rename({'vehicleType': 'vehicle_type'}, inplace=True, axis=1)
autos.rename({'abtest': 'ab_test'}, inplace=True, axis=1)
autos.rename({'gearbox': 'gear_box'}, inplace=True, axis=1)
autos.rename({'powerPS': 'power_ps'}, inplace=True, axis=1)
autos.rename({'fuelType': 'fuel_type'}, inplace=True, axis=1)
autos.rename({'nrOfPictures': 'nr_of_pictures'}, inplace=True, axis=1)
autos.rename({'postalCode': 'postal_code'}, inplace=True, axis=1)
autos.rename({'lastSeen': 'last_seen'}, inplace=True, axis=1)
```

#### Converting odometer from string to int

```python
autos.rename({'odometer': 'odometer_km'}, inplace=True, axis=1)
autos['odometer_km'] = autos['odometer_km'].str.replace('km', '')
autos['odometer_km'] = autos['odometer_km'].str.replace(',', '')
autos['odometer_km'] = autos['odometer_km'].astype(int)
```

#### Converting price from string to float

```python
autos['price'] = autos['price'].str.replace('$', '')
autos['price'] = autos['price'].str.replace(',', '')
autos['price'] = autos['price'].astype(float)
```

```python
date_crawled = autos["date_crawled"].str[:10].value_counts(normalize=True, dropna=False).sort_index()
ad_created = autos["ad_created"].str[:10].value_counts(normalize=True, dropna=False).sort_index()
last_seen = autos["last_seen"].str[:10].value_counts(normalize=True, dropna=False).sort_index()

print("-- Date Crawled --\n")
print(date_crawled)
print()
print("-- AD Created --\n")
print(ad_created)
print()
print("-- Last Seen --\n")
print(last_seen)
print()
```

#### Removing unreal registration years

```python
autos["registration_year"].describe()
```

```python
autos = autos[autos['registration_year'].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).sort_index()
```

```python
brands = autos['brand'].value_counts(normalize=True).head(10)

mean_price = {}
mean_mileage = {}

for brand in brands.index:
    mean_price[brand] = autos.loc[autos['brand']==brand,'price'].mean()
    mean_price[brand] = round(mean_price[brand], 2)

    mean_mileage[brand] = autos.loc[autos['brand']==brand,'odometer_km'].mean()
    mean_mileage[brand] = round(mean_mileage[brand], 2)
    

#print(sorted(mean_price.items(), key=lambda item: item[1], reverse=True))

price_series = pd.Series(mean_price)
mileage_series = pd.Series(mean_mileage)

df = pd.DataFrame(price_series, columns=["mean_price"])
df["mean_mileage"] = mileage_series

df

```
